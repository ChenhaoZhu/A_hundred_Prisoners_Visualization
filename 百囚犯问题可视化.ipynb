```python
# Using jumping strategy
record0 = {}
def makeRecord(record, current, card) :
    people = list(range(100))
    indi = current
    target = indi
    time = 50
    count = 1
    judge = 0
    while judge == 0 and count <= 50:
        if card[indi] != target :
            indi = card[indi]
            judge = 0
        else :
            judge = 1
        count += 1
    if judge == 0 :
        record[target] = 'Fail'
    else :
        record[target] = 'Success'
    return record



def getPercent(record) :
    count = 0
    for i in range(100) :
        if record[i] == 'Success' :
            count += 1
    return count

card0 = list(range(100))
random.shuffle(card0)
for i in range(100) :
    record = makeRecord(record0, i, card0)
print(record)
print('The times for sucess is : ', getPercent(record))


```

```python
# Using random picking strategy
record1 = {}
def makeRecord1(record, current, card) :
    people = list(range(100))
    target = current
    judge = 0
    choose = random.sample(range(0,99), 50)
    for i in choose :
        if card[i] == target :
            judge = 1
            break
        else :
            judge = 0
            
    if judge == 0 :
        record1[target] = 'Fail'
    else :
        record1[target] = 'Success'
    return record1


for i in range(100) :
    record1 = makeRecord1(record1, i, card0)
print(record1)
print('The times for sucess is : ', getPercent(record1))
```

```python
# Using spliting Strategy
record2 = {}
def makeRecord2(record, current, card) :
    people = list(range(100))
    target = current
    if current % 2 == 0 :
        choose = list(range(50))
    else :
        choose = list(range(50,100))
    judge = 0
    for i in choose :
        if card[i] == current :
            judge = 1
            break
        else :
            judge = 0
    if judge == 0 :
        record2[target] = 'Fail'
    else :
        record2[target] = 'Success'
        
    return record2


for i in range(100) :
    record2 = makeRecord2(record2, i, card0)
print(record2)
print('The times for sucess is : ', getPercent(record2))
```

```python
def randomPick(record1, card0) :
    for i in range(100) :
        record1 = makeRecord1(record1, i, card0)
    return  getPercent(record1)

def conditionalPick(record, card0) :
    for i in range(100) :
        record = makeRecord(record, i, card0)
    return getPercent(record)

def splitPick(record2, card0) :
    for i in range(100) :
        record2 = makeRecord2(record2, i, card0)
    return  getPercent(record2)


cardC = list(range(100))
list1 = []
list2 = []
list3 = []
for i in range(1000) :
    record = {}
    record1 = {}
    record2 = {}
    random.shuffle(cardC)
    list1.append(randomPick(record1, cardC))
    list2.append(conditionalPick(record, cardC))
    list3.append(splitPick(record2, cardC))
#print(list1) # ——> random pick
#print(list2) # ——> conditional pick
#print(list3) # ——> split pick
print('Done')
```

```python
# Here we plot the result
plt.figure(figsize = (10, 10))
plt.style.use('ggplot')
comp = plt.subplot(1,1,1)
comp.set_title('Hundred prisoner Question Comparision')
comp.set_xlabel('Times we test the strategy')
comp.set_ylabel('How many people find their number')
comp.scatter(list(range(1,1001)), list1, label = 'Random Pick 50 Box', color = 'deeppink')
comp.scatter(list(range(1,1001)), list2, label = 'Jumping strategy', color = 'darkblue')
comp.scatter(list(range(1,1001)), list3, label = 'Split Strategy', color = 'darkgray')
comp.legend()
```
